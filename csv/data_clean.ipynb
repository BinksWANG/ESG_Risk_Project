```python
import pandas as pd
```

```python
df = pd.read_csv("sp_500_esg_risk_ratings_original.csv")
```

```python
df.head()
```

```python
# Replace \n with empty string in all string columns
df = df.replace('\n', '', regex=True)
```

```python
df.head()
```

```python
# Remove all commas and periods from all string columns
df = df.replace('[.,]', '', regex=True)
```

```python
df.to_csv('output_file.csv', index=False)
```
